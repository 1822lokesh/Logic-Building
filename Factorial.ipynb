Factorial

```python
def factorial(num):
  factor=1
  if num < 0 :
    print("Factorial not exists negative numbers")
  elif num == 0:
    print("Factorial of 0 is 1")
  else :
    for i in range(1, num+1):
      factor = factor * i
    print(f'The factorial of {num} is {factor}')

num=int(input("Entr a num"))
res=factorial(num)
print(res)
```
